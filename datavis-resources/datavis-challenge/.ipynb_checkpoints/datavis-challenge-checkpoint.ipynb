### Analysis

- 1. Capomulin Treatment Shows Efficacy in Reducing Tumor Size:
The line plot of tumor volume over time for a specific mouse treated with Capomulin demonstrates a general reduction in tumor size throughout the study period. This suggests that Capomulin is effective in controlling or reducing tumor growth in this particular case. The trend is an encouraging indicator of the potential effectiveness of Capomulin as an anti-cancer treatment.

2. Strong Positive Correlation Between Mouse Weight and Tumor Volume:
The scatter plot and linear regression analysis reveal a strong positive correlation (approximately 0.84) between the weight of the mice and the average tumor volume for those treated with Capomulin. This implies that heavier mice tend to have larger tumor volumes. This correlation may suggest that body weight could influence tumor growth, and it might be important to consider weight when determining the dosage or effectiveness of Capomulin.

3. Variation in Tumor Volume Across Different Treatment Regimens:
The box plot of tumor volumes across different treatment regimens (Capomulin, Ramicane, Infubinol, and Ceftamin) highlights the variation in responses to these treatments. Notably, Capomulin and Ramicane show lower median tumor volumes compared to Infubinol and Ceftamin, indicating that these two regimens might be more effective in controlling tumor growth. Additionally, there are potential outliers in the Infubinol and Ceftamin groups, suggesting that some mice did not respond as well to these treatments, or there could be individual differences in response to the treatment.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import pearsonr, linregress

# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
unique_mice = combined_data["Mouse ID"].nunique()
print(f"Number of unique mice: {unique_mice}")
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = combined_data[combined_data.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]
print(duplicate_mice)
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mice_ids = combined_data[combined_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]['Mouse ID'].unique()
duplicate_mice_data = combined_data[combined_data['Mouse ID'].isin(duplicate_mice_ids)]
print(duplicate_mice_data)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_data.drop_duplicates(subset=["Mouse ID", "Timepoint"])
clean_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_unique_mice = clean_data["Mouse ID"].nunique()
print(f"Number of unique mice after cleaning: {clean_unique_mice}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.

summary_stats = clean_data.groupby("Drug Regimen").agg(mean_tumor=('Tumor Volume (mm3)', 'mean'),
                                                      median_tumor=('Tumor Volume (mm3)', 'median'),
                                                      var_tumor=('Tumor Volume (mm3)', 'var'),
                                                      std_tumor=('Tumor Volume (mm3)', 'std'),
                                                      sem_tumor=('Tumor Volume (mm3)', 'sem'))

summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

summary_statistics = combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(
    Mean="mean",
    Median="median",
    Variance="var",
    Standard_Deviation="std",
    SEM="sem"
)

summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

final_tumor = clean_data.groupby('Mouse ID').agg(last_timepoint=('Timepoint', 'max')).reset_index()
final_tumor_data = pd.merge(final_tumor, clean_data, left_on=['Mouse ID', 'last_timepoint'], right_on=['Mouse ID', 'Timepoint'])
final_tumor_data.head()
```

```python
# Define treatments
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create an empty tumor list
tumor_volumes = []

# Put treatments into a list for for loop (and later for plot labels)
final_tumor = clean_data.groupby('Mouse ID').agg(last_timepoint=('Timepoint', 'max')).reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_data = pd.merge(final_tumor, clean_data, left_on=['Mouse ID', 'last_timepoint'], right_on=['Mouse ID', 'Timepoint'])

# Calculate the IQR and determine potential outliers for each treatment
for drug in treatments:
    drug_data = final_tumor_data[final_tumor_data["Drug Regimen"] == drug]
    tumor_vol = drug_data["Tumor Volume (mm3)"]

    # Append the tumor volumes to list
    tumor_volumes.append(tumor_vol)
    
    # Calculate the quartiles and IQR
    quartiles = tumor_vol.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq

    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)

    # Add any potential outliers
    outliers = tumor_vol[(tumor_vol < lower_bound) | (tumor_vol > upper_bound)]
    print(f"{drug}'s potential outliers: {outliers}")

# Display the merged data to verify
print(f"Length of treatments: {len(treatments)}")
print(f"Length of tumor_volumes: {len(tumor_volumes)}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
plt.figure(figsize=(10, 6))
plt.boxplot(tumor_volumes, labels=treatments, flierprops={'marker': 'o', 'markerfacecolor': 'r', 'markersize': 8, 'linestyle': 'none'})
plt.title("Tumor Volume Distribution for Each Treatment Group")
plt.xlabel("Drug Regimen")
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
#Specify mouse and filter the data
capomulin_mouse_id = "l509"

capomulin_mouse_data = clean_data[(clean_data["Drug Regimen"] == "Capomulin") & 
                                     (clean_data["Mouse ID"] == capomulin_mouse_id)]

# Plot the data
plt.plot(capomulin_mouse_data["Timepoint"], capomulin_mouse_data["Tumor Volume (mm3)"], marker='o')

# Title and Labels
plt.title(f"Tumor Volume vs. Time Point for Mouse {capomulin_mouse_id} Treated with Capomulin")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")

# Display the plot
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Filter the data
capomulin_data = combined_data[combined_data["Drug Regimen"] == "Capomulin"]

# Calculate the average
capomulin_avg_tumor_vol = capomulin_data.groupby("Mouse ID").mean(numeric_only=True)["Tumor Volume (mm3)"]

# Get corresponding weights
mouse_weights = capomulin_data.groupby("Mouse ID").mean(numeric_only=True)["Weight (g)"]

# Generate scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(mouse_weights, capomulin_avg_tumor_vol, marker='o', color='blue')

# Add title and labels
plt.title("Mouse Weight vs. Average Tumor Volume for Capomulin Regimen")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

# Correct function to display the plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation, _ = pearsonr(capomulin_data['Weight (g)'], capomulin_data['Tumor Volume (mm3)'])
slope, intercept, r_value, p_value, std_err = linregress(capomulin_data['Weight (g)'], capomulin_data['Tumor Volume (mm3)'])

# Plot the correlation and regression
plt.scatter(capomulin_data['Weight (g)'], capomulin_data['Tumor Volume (mm3)'], color='blue', label='Data Points')
plt.plot(capomulin_data['Weight (g)'], slope * capomulin_data['Weight (g)'] + intercept, color='red', label='Regression Line')

# Label and title
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin Treatment: Tumor Volume vs. Weight')
plt.text(0.1, 0.9, f'Correlation: {correlation:.2f}', transform=plt.gca().transAxes, fontsize=12, color='green')
plt.legend()
plt.show()
```
